```python
import pandas as pd
import numpy as np

concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()

concrete_data.isnull().sum()

concrete_data_columns = concrete_data.columns
```

```python
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column

n_cols = predictors.shape[1] # number of predictors

#Normalizing the data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

err = np.zeros(50)


for i in range(50):
   
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30, random_state=42)

    # fit the model
    model.fit(X_train, y_train, epochs=50, verbose=2)

    y_pred = model.predict(X_test)
    err[i] = mean_squared_error(y_test, y_pred)
```

```python
#Mean error and standard deviation 
mean_err = np.mean(err)
std_err = np.std(err)
```

```python
print("The mean squared error {} and standard deviation is {}".format(  mean_err, std_err ))
```
